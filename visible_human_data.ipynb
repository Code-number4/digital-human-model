```python
# 1. Download the zip file using wget
!wget https://data.lhncbc.nlm.nih.gov/public/Visible-Human/Sample-Data/MRI-MaleFeet.zip

# 2. Unzip the file into a folder named 'male_feet_mri'
!unzip MRI-MaleFeet.zip -d male_feet_mri

# 3. List the contents to verify the download
!ls -R male_feet_mri
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from glob import glob

# 1. Define the path where you unzipped the files
data_path = 'male_feet_mri/'

# 2. Get a list of all image files (adjust extension if they are .png or .tif)
# Using glob helps find files even if they are in subdirectories
image_files = sorted(glob(os.path.join(data_path, '**/*.jpg'), recursive=True))

if not image_files:
    print("No images found. Check the file extension or the data_path.")
else:
    # 3. Set up a grid to view the first 9 slices
    plt.figure(figsize=(15, 15))

    for i in range(min(9, len(image_files))):
        plt.subplot(3, 3, i + 1)
        img = mpimg.imread(image_files[i])
        plt.imshow(img, cmap='gray')
        plt.title(f"Slice: {os.path.basename(image_files[i])}")
        plt.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 1. Define the specific files based on your zip extraction
data_path = 'male_feet_mri/'
file_names = ['m_vm7473.t1.png', 'm_vm7473.t2.png', 'm_vm7473.pd.png']
scan_types = ['T1-Weighted', 'T2-Weighted', 'Proton Density (PD)']

# 2. Set up a wide 1x3 grid
plt.figure(figsize=(18, 6))

for i, file_name in enumerate(file_names):
    img_path = os.path.join(data_path, file_name)

    # 3. Read and display the image
    img = mpimg.imread(img_path)
    plt.subplot(1, 3, i + 1)
    plt.imshow(img, cmap='gray')

    # 4. Add the title showing the scan type and filename
    plt.title(f"{scan_types[i]}\n({file_name})", fontsize=14)
    plt.axis('off')

plt.tight_layout()
plt.show()
```
